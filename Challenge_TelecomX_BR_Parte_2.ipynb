```python
one_hot.get_feature_names_out(colunas)
```

Balanceando os dados com SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
x_balanceado, y_balanceado = oversample.fit_resample(x, y)
```

```python
import pandas as pd

# Como y_balanceado √© um array √© necess√°rio tranform√°-lo para Series p/ usar value_counts()
y_balanceado_series = pd.Series(y_balanceado)

# Calcula os valores dos dados balanceados
proporcao_y_balanceado = y_balanceado_series.value_counts()

# Mostra a propor√ß√£o das classes
print("Propor√ß√£o de evas√£o ap√≥s balanceamento:")
print(proporcao_y_balanceado)

# Mostra a propor√ß√£o como porcentagem
print("\nPropor√ß√£o em porcentagem:")
print(y_balanceado_series.value_counts(normalize=True) * 100)
```

# ü§ñ Modelagem Preditiva

Separa√ß√£o de Dados 80/20

```python
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x_balanceado, y_balanceado, test_size=0.2, random_state=42)
```

Cria√ß√£o de modelos

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

Modelo √Årvore de Decis√£o, m√©tricas e matriz de confus√£o

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

decision_tree_model = DecisionTreeClassifier(random_state=42)

decision_tree_model.fit(x_treino, y_treino)

y_pred_dt = decision_tree_model.predict(x_teste)

accuracy_dt = accuracy_score(y_teste, y_pred_dt)
precision_dt = precision_score(y_teste, y_pred_dt)
recall_dt = recall_score(y_teste, y_pred_dt)
f1_dt = f1_score(y_teste, y_pred_dt)
conf_matrix_dt = confusion_matrix(y_teste, y_pred_dt)

print("M√©tricas de avalia√ß√£o para o modelo de √Årvore de Decis√£o:")
print(f"Acur√°cia: {accuracy_dt:.4f}")
print(f"Precis√£o: {precision_dt:.4f}")
print(f"Recall: {recall_dt:.4f}")
print(f"F1-score: {f1_dt:.4f}")
print("\nMatriz de Confus√£o:")
print(conf_matrix_dt)
```

Modelo de Random Forest, m√©tricas e matriz de confus√£o

```python
random_forest_model = RandomForestClassifier(random_state=42)

random_forest_model.fit(x_treino, y_treino)

y_pred_rf = random_forest_model.predict(x_teste)

accuracy_rf = accuracy_score(y_teste, y_pred_rf)
precision_rf = precision_score(y_teste, y_pred_rf)
recall_rf = recall_score(y_teste, y_pred_rf)
f1_rf = f1_score(y_teste, y_pred_rf)
conf_matrix_rf = confusion_matrix(y_teste, y_pred_rf)

print("M√©tricas de avalia√ß√£o para o modelo de Random Forest:")
print(f"Acur√°cia: {accuracy_rf:.4f}")
print(f"Precis√£o: {precision_rf:.4f}")
print(f"Recall: {recall_rf:.4f}")
print(f"F1-score: {f1_rf:.4f}")
print("\nMatriz de Confus√£o:")
print(conf_matrix_rf)
```

Avalia√ß√£o dos modelos

```python
print("Compara√ß√£o das m√©tricas de avalia√ß√£o:")
print("-" * 40)
print(f"{'M√©trica':<15} | {'√Årvore de Decis√£o':<20} | {'Random Forest':<15}")
print("-" * 40)
print(f"{'Acur√°cia':<15} | {accuracy_dt:<20.4f} | {accuracy_rf:<15.4f}")
print(f"{'Precis√£o':<15} | {precision_dt:<20.4f} | {precision_rf:<15.4f}")
print(f"{'Recall':<15} | {recall_dt:<20.4f} | {recall_rf:<15.4f}")
print(f"{'F1-score':<15} | {f1_dt:<20.4f} | {f1_rf:<15.4f}")
print("-" * 40)
```

An√°lise Cr√≠tica:
O modelo Random Forest teve um desempenho superior ao modelo de √Årvore de Decis√£o.

Ele apresentou m√©tricas de acur√°cia, precis√£o, recall e F1-score mais altas.

Para a predi√ß√£o de evas√£o, a precis√£o e o recall o se mostraram condizentes com a matriz de confu√ß√£o.

O Random Forest demonstrou um melhor equil√≠brio entre essas m√©tricas, conforme indicado pelo F1-score superior.
